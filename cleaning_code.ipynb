```python
import glob
import pandas as pd
import os
```

```python
folder_path = r"D:\_Computer_Science_\DEPI\The Final Project\Data_Climate_Change"
years_str=["2021-2022","2022-2023","2023-2024"]
community_names = ["Agriculture", "Sustainable Buildings", "Renewable Energy"]
```

# Removing Headers

```python
# function to remove the metadata from everyearfile
def remove_first_n_lines(file_path, n=9):
    lines = []
    with open(file_path, "r") as f:
        lines = f.readlines()[n:]
    with open(file_path, "w") as f:
        f.writelines(lines)
```

```python
for file_path in glob.glob(f"{folder_path}/**/*.csv", recursive=True):
    remove_first_n_lines(file_path)
```

# Removing Nulls

```python
def remove_nulls(file_path):
    df = pd.read_csv(file_path)
    df = df[df.iloc[:, -1] != -999]
    df.to_csv(file_path, index=False)
```

# Rename

```python
def rename_files(path):
    for file_path in glob.glob(f"{path}/*.csv", recursive=True):
        df = pd.read_csv(file_path)
        df.to_csv(f"{path}/{df.columns[-1]}.csv", index=False)
```

```python
for i in years_str:
    for c in community_names:
    rename_files(f"{folder_path}/{c}/{i}")
```

# Append

## append the common in the 3 years

```python
cols = {}
for community in community_names:
    comm_cols = set()
    for year in years_str:
        year_cols = set()
        for file_name in glob.glob(f"{folder_path}/{community}/{year}/*.csv", root_dir=folder_path, recursive=True):
            year_cols.add(file_name.split('\\')[-1].split('.')[0])
        if len(comm_cols) == 0:
            comm_cols = year_cols
        else:
            comm_cols = comm_cols.intersection(year_cols)
    cols[community] = list(comm_cols)
cols
```

```python
for community, col_names in cols.items():
    for col in col_names:
        dataframes = []
        for file_path in glob.glob(f"{folder_path}/{community}/*/{col}.csv", root_dir=folder_path, recursive=True):
            dataframes.append(pd.read_csv(file_path))
            os.remove(file_path)
        df = pd.concat(dataframes)
        df.to_csv(f"D:/_Computer_Science_/DEPI/The Final Project/appended/{community}/{col}.csv")
```

```python
def Merge(community):
    df_32=None
    df_104=None
    for cur_file in glob.glob(f"{folder_path}/{community}/*.csv"):
        test=pd.read_csv(cur_file).reset_index(drop=True)
        
        if df_32 is None and (len(test)==32940 or len(test)==32940*2 or len(test)==32940*3):
            df_32=test
        elif df_104 is None and (len(test)==104310 or len(test)==104310*2 or len(test)==104310*3):
            df_104 =test
        else :
            if df_32 is not None and (len(test) == 32940 or len(test)==32940*2 or len(test)==32940*3):
                df_32 = pd.merge(df_32, test, how='outer').reset_index(drop=True)
            elif df_104 is not None and (len(test) == 104310 or len(test)==104310*2 or len(test)==104310*3):
                df_104 = pd.merge(df_104, test, how='outer').reset_index(drop=True)
    
    if df_32 is not None:
        df_32.to_csv(f"D:/_Computer_Science_/DEPI/The Final Project/merged/{community}_df_32.csv", index=False)
    if df_104 is not None:
        df_104.to_csv(f"D:/_Computer_Science_/DEPI/The Final Project/merged/{community}_df_104.csv", index=False)
```
